```python
import numpy as np
import csv
import pandas as pd
import matplotlib.pyplot as plt


# Read in motor data
motor_data_folder = "tests/System Tests/PCBTesting_MaxTorque"
motor_id = 6
Kp = 40
Ki = 30
v = 20
TC = 2000
test_name = f"single_motor_ID{motor_id}_Kp{Kp}_Ki{Ki}_v{v}_TC{TC}"

# with open(f"{motor_data_folder}/{test_name}", 'r') as f:
# 	reader = csv.reader(f)
# 	for line in reader:
# 		print(line)

motor_data = pd.read_csv(f"{motor_data_folder}/{test_name}")
max_motor_torque = motor_data["torque"].max()

# Read in fts data
fts_file = "C:/Users/jlimk/Documents/4-10-24_motor_tests/screw6_max_torque_current_2000_Kp40_Ki30_speed20.csv"
fts_data = pd.read_csv(fts_file, skiprows=list(range(6)))
max_sensor_torque = fts_data[' Tz '].max()

fig, axes = plt.subplots(1, 2, figsize=(12, 6))

fig.suptitle("Screw 6, Kp = 40, Speed = 20")
axes[0].plot(motor_data["angular speed"])
axes[0].plot(motor_data["torque"])
axes[0].axhline(max_motor_torque)
axes[0].legend(["ang vel", "torque"])
axes[0].set_xlabel("time")

axes[1].plot(fts_data[' Tz '])
axes[1].axhline(max_sensor_torque)
axes[1].legend(['Sensor Torque'])

plt.style.use('bmh')
plt.tight_layout()
plt.show()

measured_torque_constant = max_sensor_torque / max_motor_torque
print(f"Max Motor Torque = {max_motor_torque}")
print(f"Max Sensor Torque = {max_sensor_torque}")
print(f"Torque Constant = {measured_torque_constant}")
```
